```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import pandas as pd
import requests
import json

# List of deals data
request = requests.get('https://www.cheapshark.com/api/1.0/deals')
deal_data = request.json()

# game Data
req = requests.get('https://www.cheapshark.com/api/1.0/stores')
store_info = request.json()

# info about the stor


# Create a json file call "deals_data" having the data

#with open('deals_data.json', 'w') as f:
    #json.dump(deals_data, f)
```

```python
# Dataframe 

#deal_df = pd.DataFrame(columns = ['Store_ID', 'Savings', 'Deal_Rating'])
deal_df = pd.DataFrame(columns = ['Store Name', 'Store ID', 'Savings Percentage', 'Deal Rating'])
```

```python
#with open('deals_data.json', 'r') as f:
    
#my_data = json.load(f)
for t in deal_data:
    #print('\n', t['title'])
        
    store_id = int(t["storeID"])
    #print(f'store id: {store_id}')
        
    savings = float(t["savings"])
    #print(f'saving: {savings}')
        
    deal_rating = float(t["dealRating"])
    #print(f'deal rating: {deal_rating}')
    
    for k in store_info:
        #print(x["storeName"])
        if k["storeID"] == t["storeID"]:
            
            store_name = k["storeName"]
        
            deal_df.loc[len(deal_df.index)] = [store_name, store_id, savings, deal_rating]

        
        
    #deal_df.loc[len(deal_df.index)] = [store_id, savings, deal_rating]
        
        #print('\n', t)
        
   
```
